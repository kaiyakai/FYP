```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('accuracy_back_prob_change.csv')
df['Unnamed: 0'] = 100 - df['Unnamed: 0']
df = df.rename(columns={'Unnamed: 0':'hedge ratio', '1':'segment length=2'}).sort_values(by='hedge ratio', ascending=False).reset_index(drop=True)
```

```python
df.plot(x='hedge ratio', figsize=(8,6))
plt.ylabel('accuracy')
plt.title('accuracy of displacement prediction of \n under different hedge ratio')
plt.savefig('../image/effect_of_hedge_ratio.png')
```

```python
plt.figure(figsize=(8,6))
for i in [23,49,77,86]:
    df = pd.read_csv(f'accuracy_n_change_{i}.csv')
    df['Unnamed: 0'] = 100 - df['Unnamed: 0']
    df = df.rename(columns={'Unnamed: 0':'hedge_ratio'}).sort_values(by='hedge_ratio', ascending=False).reset_index(drop=True)
    plt.plot(np.array(range(1,len(df.to_numpy()[0][1:])+1)) * 2, df.to_numpy()[0][1:])
plt.title('accuracy of displacement prediction under different segment length n')
plt.xlabel('segment length n')
plt.ylabel('accuracy')
plt.legend([f'hedge ratio = {100-i}' for i in [23,49,77,86]])
plt.savefig('../image/effect_of_segment_length.png')
```

```python
from tqdm import tqdm
import os
import numpy as np

sn_list = []
real_back_prob_mean_list = []
for sn in tqdm(range(99,0,-1), ncols=70):
    frame_dir = '../scene_data/{}'.format(sn)

    scene_dict = {}
    real_back_prob_list = []
    for file_name in os.listdir(frame_dir):
        if file_name[0] != '.':
            num, verti_pos, hori_pos, real_back_prob = file_name.replace('.jpg', '').split('_')
            real_back_prob_list.append(int(real_back_prob))
            scene_dict[int(num)] = {'file_name': file_name, 'num':num, 'verti_pos':int(verti_pos), 'hori_pos':int(hori_pos)}
    real_back_prob_mean = int(np.mean(real_back_prob_list))
    sn_list.append(sn)
    real_back_prob_mean_list.append(real_back_prob_mean)
```
